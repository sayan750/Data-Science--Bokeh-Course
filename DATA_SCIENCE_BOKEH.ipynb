**Bokeh**

**Bokeh is a Python library for creating interactive and visually appealing data visualizations in web browsers. It provides a powerful set of tools for creating interactive plots, dashboards, and data applications that can be easily deployed on the web.**

**Bokeh supports a wide range of visualization types, including line plots, scatter plots, bar charts, heatmaps, and more. It also offers several advanced features, such as linked brushing and linking multiple plots together to create interactive dashboards.**

**One of the unique features of Bokeh is its ability to create interactive widgets that allow users to interact with the data in real-time. For example, sliders and dropdown menus can be used to filter or highlight specific data points, while hover tooltips can provide additional information about individual data points.**

**Bokeh is a popular tool in the data science community due to its ease of use, powerful capabilities, and extensive documentation. It can be used in a variety of applications, from exploratory data analysis to creating production-ready data applications.**

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.iris import flowers
```

```python
flowers
```

```python
output_file('test.html')

p = figure(title = 'Test Flower')
p.xaxis.axis_label = 'petal_length'
p.yaxis.axis_label = 'petal_width'
p.circle(flowers['petal_length'], flowers['petal_width'])
show(p)
```

```python
p = figure(title = 'Test Flower')
p.xaxis.axis_label = 'petal_length'
p.yaxis.axis_label = 'petal_width'
p.line(flowers['petal_length'], flowers['petal_width'])
show(p)
```

```python
x = [1,2,3,4,5,6]
y = [4,6,7,29,5,0,1]
output_file('line.html')
d = figure(title = 'line_graph')
d.line(x,y)
show(d)
```

```python
x = [1,2,3,4,5,6]
y = [4,6,7,29,5,0,1]
output_file('line.html')
d = figure(title = 'line_graph')
d.scatter(x,y, size = 20, color = 'pink', legend_label = 'This is my data point')
show(d)
```
